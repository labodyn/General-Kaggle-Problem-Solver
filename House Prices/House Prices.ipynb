```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
# Read in data
df = pd.read_csv('data/train.csv')
df_test = pd.read_csv('data/test.csv')
df_full = pd.concat([df, df_test])

# Define important variable names of the data set
outcome_str = 'SalePrice'
id_str = 'Id'
```

# Model phase

```python
# Split in train and test set. For submissions, all data is used
df_train = df.sample(frac=0.8,random_state=42)
df_cv = df.drop(df_train.index)
```

```python
# Define the metric: root mean squared logarithmic error
def rmsle(y_true, y_pred):
    return sqrt(mean_squared_error(np.log(y_true), np.log(y_pred)))
```

```python
# Get missing features
all_feats = df_full.columns
missing_feat = [feat for feat in df_full.columns if sum(df[feat].isnull())]
numeric_feats = df_full.dtypes[df_full.dtypes != "object"].index
cat_feats = df_full.dtypes[df_full.dtypes == "object"].index
```

```python
# Feature selection
def select_features(df):
    selected_features = ['LotArea']#, 'Neighborhood']
    df_selected = df[selected_features]
    df_selected = df_selected.join(pd.get_dummies(df['Neighborhood']))
    return df_selected
```

```python
# Prepare datasets to train on
X_train = select_features(df_train)
X_cv = select_features(df_cv)
y_train = df_train[outcome_str]
y_cv = df_cv[outcome_str]
```

*** Models ***

```python
def evaluate_model(model_str):
    model = eval(model_str)
    model.fit(X_train, y_train)
    y_train_pred = model.predict(X_train)
    y_cv_pred = model.predict(X_cv)
    print('--- {} ---'.format(model_str))
    print('Root mean squared logaritmic error on training data:         {:1.3f}'.format(rmsle(y_train, y_train_pred)))
    print('Root mean squared logaritmic error on cross-validation data: {:1.3f}\n'.format(rmsle(y_cv, y_cv_pred)))
```

```python
# Define models as a string
models = ['LinearRegression()',
          'RandomForestRegressor()',
          'GradientBoostingRegressor()',
          'GradientBoostingRegressor(n_estimators=500, max_depth=2)'
         ]

for model_str in models:
    evaluate_model(model_str)
```

# Submission phase

```python
# Decide upon final model
final_model = GradientBoostingRegressor()
```

```python
# Prepare datasets to train on
X = select_features(df)
X_to_submit = select_features(df_test)
y = df[outcome_str]

# Fit model on full dataset
final_model.fit(X, y)

# Predict
y_to_submit = final_model.predict(X_to_submit)
```

```python
# Write out submission
df_submission = pd.DataFrame(data=y_to_submit, index=df_test[id_str], columns=[outcome_str])
df_submission.to_csv('my_submission.csv')
```
